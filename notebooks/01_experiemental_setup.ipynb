# Kísérleti beállítások és módszertan

Ebben a beadandóban a **SARSA** és a **Q-learning** megerősítéses tanulási algoritmusokat
hasonlítom össze két klasszikus környezeten:

- **FrozenLake** – diszkrét állapot- és akciótér
- **CartPole** – folytonos állapottér diszkretizálva

A cél annak vizsgálata, hogy az **on-policy** (SARSA) és **off-policy** (Q-learning)
megközelítés hogyan viselkedik különböző környezeti feltételek mellett.


## Vizsgált algoritmusok

### SARSA (on-policy)
A SARSA algoritmus a frissítés során a ténylegesen végrehajtott következő akciót
használja, így a tanult értékfüggvény közvetlenül a követett policy-hez kötődik.

\[
Q(s,a) \leftarrow Q(s,a) + \alpha (r + \gamma Q(s', a') - Q(s,a))
\]

### Q-learning (off-policy)
A Q-learning algoritmus a következő állapotban elérhető maximális Q-értéket
használja, függetlenül a ténylegesen végrehajtott akciótól.

\[
Q(s,a) \leftarrow Q(s,a) + \alpha (r + \gamma \max_a Q(s', a) - Q(s,a))
\]


## Környezetek

### FrozenLake
- Diszkrét állapot- és akciótér
- Két változat:
  - determinisztikus (`is_slippery = False`)
  - sztochasztikus (`is_slippery = True`)

### CartPole
- Folytonos állapottér (pozíció, sebesség, szög, szögsebesség)
- Tabuláris algoritmusok miatt diszkretizálva
- A diszkretizáció granularitása befolyásolja a tanulás stabilitását
