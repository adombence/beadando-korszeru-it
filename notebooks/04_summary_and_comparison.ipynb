```python
import sys
from pathlib import Path
import pandas as pd

# Add parent directory to path to import utils
sys.path.insert(0, str(Path.cwd().parent))

import matplotlib.pyplot as plt
from util import load_logs, moving_average

sarsa = load_logs("FrozenLake-v1", "SARSA")
qlearning = load_logs("FrozenLake-v1", "Q-learning")
```

```python
def summarize(df: pd.DataFrame, metric: str):
    return df.groupby("seed")[metric].mean().mean(), df.groupby("seed")[metric].mean().std()

summary = pd.DataFrame({
    "Environment": ["FrozenLake", "CartPole"],
    "SARSA (mean ± std)": [
        summarize(sarsa := load_logs("FrozenLake-v1", "SARSA"), "success"),
        summarize(load_logs("CartPole-v1", "SARSA"), "length")
    ],
    "Q-learning (mean ± std)": [
        summarize(load_logs("FrozenLake-v1", "Q-learning"), "success"),
        summarize(load_logs("CartPole-v1", "Q-learning"), "length")
    ],
})

summary
```
